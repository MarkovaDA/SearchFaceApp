```python
import sys
import dlib
from skimage import io
#  Histogram of Oriented Gradients
#https://algotravelling.com/ru/%D0%BC%D0%B0%D1%88%D0%B8%D0%BD%D0%BD%D0%BE%D0%B5-%D0%BE%D0%B1%D1%83%D1%87%D0%B5%D0%BD%D0%B8%D0%B5-%D1%8D%D1%82%D0%BE-%D0%B2%D0%B5%D1%81%D0%B5%D0%BB%D0%BE-4/
face_detector = dlib.get_frontal_face_detector()
file_name = '/Users/daryamarkova/AnacondaProjects/face_pictures/face3.jpg' # 
image = io.imread(file_name)

#io.imshow(image)

detected_faces = face_detector(image, 1)
#         file
#[(167, 150) (322, 305)]
for i, face_rect in enumerate(detected_faces):
    x1 = face_rect.left()
    x2 = face_rect.right()
    y1 = face_rect.top()
    y2 = face_rect.bottom()

#face_img = image[x1:x2, y1:y2]
face_img = image[y1:y2, x1:x2]
io.imshow(face_img)
#    
io.imsave('/Users/daryamarkova/AnacondaProjects/face_pictures/short_face.jpg', face_img) 
```
